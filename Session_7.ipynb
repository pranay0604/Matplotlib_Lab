Multi-Variate
---

- Scatter

- Facetgrid

- Pairplot

- PairGrid

```python
sns.scatterplot(data=df,x='age',y='salary',hue='department',style='gender')
```

```python
sns.scatterplot(data=df,x='age',y='salary',style='gender')
```

```python
sns.pairplot(df[['age','work_hours_per_week','salary','experience_years','gender']],hue='gender')
```

```python
sns.pairplot(
    data=df,
    vars=['age','work_hours_per_week','salary','experience_years'],
    hue='gender'
)
```

```python
g = sns.FacetGrid(df,hue='gender')
g.map(sns.scatterplot,'age','salary')
```

```python
g = sns.FacetGrid(df, hue='gender', col='gender')
g.map(sns.scatterplot, 'experience_years', 'salary')
```

```python
g = sns.FacetGrid(df, hue='gender')
g.map(sns.scatterplot, 'experience_years', 'salary')
g.add_legend()
```

```python
g = sns.PairGrid(df[['age','work_hours_per_week','salary','experience_years','gender']], hue='gender')
g.map(sns.scatterplot)
g.add_legend()
```

```python
g = sns.FacetGrid(df[['age','salary','gender']], hue='gender')
g.map(sns.histplot, 'salary')
g.add_legend()
```

```python
g = sns.FacetGrid(df, col='gender', hue='gender')
g.map(sns.histplot, 'salary')
g.add_legend()
```

```python
sns.displot(
    data=df,
    x='salary',
    col='gender',
    kind='hist'
)
```
