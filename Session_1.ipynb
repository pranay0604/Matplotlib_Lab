The main parts of Matplotlib
---

- window

- axis -> this is the workplace of the matplotlib

- x/y axis -> it shows x label and y label information

- plt.plot() -> connectivity to the points inside axes

- plt.show() -> display of the plot

- plt.title() -> title of the plot

- we can find hidden logics and patterns and insights

```python
import matplotlib.pyplot as plt
```

```python
plt.plot() # inside plot insert x and y values
```

```python
hours = [4,6,2,6,8,2,9,4]
marks = [41,61,21,69,88,29,96,49]

plt.plot(hours,marks,marker='o')
```

```python
hours = [4,6,2,6,8,2,9,4]
marks = [41,61,21,69,88,29,96,49]

plt.plot(hours,marks,marker='o')
plt.title("Hours vs Marks")
plt.show()
```

```python
hours = [4,6,2,6,8,2,9,4]
marks = [41,61,21,69,88,29,96,49]

plt.plot(hours,marks,marker='o')
plt.title("Hours vs Marks",fontdict={'fontsize':30, 'color':'red'})
plt.show()
```

```python
hours = [4,6,2,6,8,2,9,4]
marks = [41,61,21,69,88,29,96,49]

plt.plot(hours,marks,marker='o')
plt.title("Hours vs Marks",fontdict={'fontsize':30, 'color':'green'})
plt.xlabel("Hours",fontdict={'fontsize':20, 'color':'orange'})
plt.ylabel("Marks",fontdict={'fontsize':20, 'color':'red'})
plt.show()
```

```python
hours = [4,6,2,6,8,2,9,4]
marks = [41,61,21,69,88,29,96,49]

plt.plot(hours,marks,marker='o',label='Hours vs Marks')
plt.title("Hours vs Marks",fontdict={'fontsize':30, 'color':'green'})
plt.xlabel("Hours",fontdict={'fontsize':20, 'color':'orange'})
plt.ylabel("Marks",fontdict={'fontsize':20, 'color':'red'})
plt.legend()
plt.show()
```

```python
hours = [4,6,2,6,8,2,9,4]
mob_time = [10,5,12,4,3,11,2,9]
marks = [41,61,21,69,88,29,96,49]

plt.plot(hours,marks,marker='o',label='Hours vs Marks')
plt.plot(mob_time,marks,marker='d',label='Mobile Time')

plt.title("Hours vs Marks",fontdict={'fontsize':30, 'color':'green'})
plt.xlabel("Hours",fontdict={'fontsize':20, 'color':'orange'})
plt.ylabel("Marks",fontdict={'fontsize':20, 'color':'red'})

plt.legend()
plt.show()
```

```python
hours = [4,6,2,6,8,2,9,4]
mob_time = [10,5,12,4,3,11,2,9]
marks = [41,61,21,69,88,29,96,49]

plt.plot(hours,marks,marker='o',ms=10,mec='red',mfc='green',label='Hours vs Marks')
plt.plot(mob_time,marks,marker='d',ms=10,mec='green',mfc='red',label='Mobile Time')

plt.title("Hours vs Marks",fontdict={'fontsize':30, 'color':'green'})
plt.xlabel("Hours",fontdict={'fontsize':20, 'color':'orange'})
plt.ylabel("Marks",fontdict={'fontsize':20, 'color':'red'})

plt.legend()
plt.show()
```
