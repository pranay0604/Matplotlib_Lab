```python
import pandas as pd


data = {
    'empid': [101, 102, 103, 104, 105],
    'name': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'David Lee', 'Eve Davis'],
    'salary': [60000.0, 75000.0, 65000.0, 80000.0, 70000.0],
    'age': [30, 45, 35, 40, 28],
    'gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'department': ['HR', 'Engineering', 'Sales', 'Engineering', 'IT'],
    'joining date': ['2015-05-03', '2010-11-09', '2019-02-12', '2017-09-08', '2020-02-15'],
    'experience': [5, 15, 6, 10, 3],
    'performance rating': [3, 4, 3, 4, 5]
}


df = pd.DataFrame(data)

df
```

Boxplot :-

          Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                       |-----:-----|
       o      |--------|     :     |--------|    o  o
                       |-----:-----|
     flier             <----------->            fliers
                            IQR

```python
plt.boxplot(df['salary'],showmeans=True)
```

```python
plt.boxplot(df['salary'],showmeans=True,tick_labels=[df['experience']])
```

```python
df.boxplot('salary','experience')
```

```python
groups = [df[df['experience'] == i]['salary'] for i in df['experience'].unique()]

plt.boxplot(groups, showmeans=True, tick_labels=df['experience'].unique())
plt.show()
```
