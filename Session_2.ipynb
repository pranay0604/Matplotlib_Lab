```python
x = range(4,40,3)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,marker='o')
plt.plot(x,z,marker='o')
plt.show()
```

```python
x = range(4,40,3)
y = np.sin(x)
z = np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y,marker='o',color='green')
plt.subplot(1,2,2)
plt.plot(x,z,marker='o',color='orange')
plt.show()
```

```python
x = range(4,40,3)
y = np.sin(x)
z = np.cos(x)
x1 = np.tan(x)
x2 = np.log(x)

plt.subplot(2,2,1)
plt.plot(x,y,marker='o',color='green')

plt.subplot(2,2,2)
plt.plot(x,z,marker='o',color='orange')

plt.subplot(2,2,3)
plt.plot(x,x1,marker='o',color='green')

plt.subplot(2,2,4)
plt.plot(x1,x2,marker='o',color='orange')

plt.show()
```
