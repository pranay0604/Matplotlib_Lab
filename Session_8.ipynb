SUBPLOTS (Matplotlib)
=====================

• Subplots are used to display multiple plots in a single figure.

• They help compare different graphs side by side.

• Each subplot is arranged in rows and columns.

• Commonly used in data visualization for better analysis.

Syntax :
        plt.subplot(nrows, ncols, index)

Another Method (Using subplots function):

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x,y)

```python
fig, axes = plt.subplots(2,3)
# fig = window , axes = positions

axes[0,1].plot([7,4,5],[1,2,3])
```

```python
fig, axes = plt.subplots(2,3)
# fig = window , axes = positions

axes[0,1].plot([7,4,5],[1,2,3])

axes[1,2].plot([7,4,5],[1,2,3])
```

```python
fig, axes = plt.subplots(2,3)
# fig = window , axes = positions

axes[0,1].plot([7,4,5],[1,2,3])

plt.title('x line')
plt.xlabel('x label')
plt.ylabel('y label')
axes[1,2].plot([7,4,5],[1,2,3])
plt.tight_layout()
```
