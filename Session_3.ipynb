```python
1) Normal Distribution

2) Left Skewed Distribution

3) Right Skewed Distribution

4) Equal Distribution
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skewnorm

x = np.linspace(-5, 5, 1000)

y_normal = norm.pdf(x, 0, 1)

y_left = skewnorm.pdf(x, -5, loc=1.5, scale=2)

y_right = skewnorm.pdf(x, 5, loc=-1.5, scale=2)

y_equal = np.where((x > -3) & (x < 3), 0.2, 0)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(x, y_normal, color='blue'); axes[0, 0].set_title("Normal Distribution")
axes[0, 1].plot(x, y_left, color='red'); axes[0, 1].set_title("Left Skewed (Negative)")
axes[1, 0].plot(x, y_right, color='green'); axes[1, 0].set_title("Right Skewed (Positive)")
axes[1, 1].plot(x, y_equal, color='purple'); axes[1, 1].set_title("Equal (Uniform) Distribution")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
n_samples = 1000

normal_data = np.random.normal(0, 1, n_samples)

left_skewed_data = np.random.beta(7, 2, n_samples)

right_skewed_data = np.random.exponential(1, n_samples)

uniform_data = np.random.uniform(0, 1, n_samples)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))


def plot_hist(ax, data, title, color):
    ax.hist(data, bins=30, color=color, edgecolor='black', alpha=0.7)
    ax.set_title(title)
    ax.grid(axis='y', alpha=0.3)

plot_hist(axes[0, 0], normal_data, 'Normal Distribution', 'skyblue')
plot_hist(axes[0, 1], left_skewed_data, 'Left Skewed (Negative)', 'salmon')
plot_hist(axes[1, 0], right_skewed_data, 'Right Skewed (Positive)', 'lightgreen')
plot_hist(axes[1, 1], uniform_data, 'Equal (Uniform) Distribution', 'orchid')

plt.tight_layout()
plt.show()
```
